```python
import pandas as pd
```

```python
df = pd.read_csv('data.csv', index_col=0)
```

```python
def number_of_words(array):
    array_num=[len(string.split()) for string in array]
    return array_num
```

```python
nwords=number_of_words(df.song_title.to_numpy())
y=df['mode'].to_numpy()
df_new=df.drop(columns=['song_title', 'artist', 'mode'])
df_new['title_nwords']=nwords
```

```python
x_features=df_new.to_numpy()
```

```python
from sklearn.model_selection import train_test_split

x_train, x_, y_train, y_ = train_test_split(x_features, y, test_size=0.40, random_state=1)

x_cv, x_test, y_cv, y_test = train_test_split(x_, y_, test_size=0.50, random_state=1)

# Delete temporary variables
del x_, y_
```

```python
import matplotlib.pyplot as plt 

fig, ax = plt.subplots()

ax.scatter(x_train[:,0], y_train, label='train')
ax.scatter(x_test[:,0], y_test, label='test')
ax.scatter(x_cv[:,0], y_cv, label='CV')
ax.legend()
```

Feature scaling

```python
from sklearn.preprocessing import StandardScaler, PolynomialFeatures

scaler_linear = StandardScaler()

X_train_scaled = scaler_linear.fit_transform(x_train)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
tf.random.set_seed(1234)  # applied to achieve consistent results
model=Sequential(
    [
        Dense(2, activation='relu', name='L1'),
        Dense(1, activation='linear', name='Out')
    ]
)
```

```python
model.compile(
    loss=tf.keras.losses.BinaryCrossentropy(from_logits=True),
    optimizer=tf.keras.optimizers.Adam(0.01)
)
```

```python

model.fit(
    X_train_scaled,y_train,
    epochs=200
)
```
